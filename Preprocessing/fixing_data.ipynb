```python
balekembang
```

```python
cacalan=pd.read_csv("annotated_pantai cacalan.csv")
```

```python
cacalan
```

```python
pulau_merah=pd.read_csv("annotated_pantai pulau merah.csv")
```

```python
goa_cina=pd.read_csv("annotated_pantai goa cina.csv")
```

```python
df=pd.concat([klayar,asmara,pantai_kelapa,balekembang,cacalan,pulau_merah,goa_cina])
```

```python
df=df.dropna()
```

```python
df
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('sentiment').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('aspect').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.to_csv('pantai_terlabeli.csv')
```

```python
df_dropped = df.drop(columns=['sentiment','aspect'])
```
