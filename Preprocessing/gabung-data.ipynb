```python
import pandas as pd
import os

# List of CSV files with lowercase names
csv_files = [
    'pantai balekembang.csv',
    'pantai klayar.csv',
    'pantai papuma.csv',
    'pantai goa cina.csv',
    'pantai ngliyep.csv',
    'pantai pulau merah.csv',
    'pantai kelapa.csv',
    'pantai prigi.csv',
    'pantai pasir putih remen tuban.csv',
    'pantai cacalan.csv',
    'pantai teluk asmara.csv'
]

# List to hold dataframes
dataframes = []

# Read each CSV file, add the 'nama_pantai' column, and append to the list
for file in csv_files:
    file_path = file
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        # Extract the beach name from the file name
        beach_name = file.replace('pantai ', '').replace('.csv', '')
        # Add the 'nama_pantai' column
        df['nama_pantai'] = beach_name
        # Drop the specified columns if they exist
        df = df.drop(columns=['sentimen', 'Unnamed: 4', 'Unnamed: 5'], errors='ignore')
        dataframes.append(df)
    else:
        print(f"File {file} not found.")

# Concatenate all dataframes
merged_df = pd.concat(dataframes, ignore_index=True)

```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Count the occurrences of each beach name
beach_counts = merged_df['nama_pantai'].value_counts()

# Create a bar graph
plt.figure(figsize=(10, 6))
beach_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Entries per Beach')
plt.xlabel('Beach Name')
plt.ylabel('Number of Entries')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the plot as an image file
plt.savefig('beach_entries_bargraph.png')

# Show the plot
plt.show()
```

```python
# Save the merged dataframe to a new CSV file
merged_df.to_csv('pantai_gabungan_unclean.csv', index=False)
```
