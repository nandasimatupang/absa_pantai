```python
import pandas as pd
import re
import os
```

```python
file_name = 'teluk asmara.csv'

data=pd.read_csv("Data/"+file_name)
```

```python
data
```

```python
data_baru=data.dropna(subset='wiI7pd')
data_baru.drop(columns=['NBa7we src', 'RfnDt','dSlJg',	'znYl0'	,'dSlJg 2',	'znYl0 2'], inplace=True)
```

```python
# Fungsi untuk menghapus emoji menggunakan regex
def remove_emoji(text):
    emoji_pattern = re.compile(
        "[" \
        u"\U0001F600-\U0001F64F"  # Emotikon
        u"\U0001F300-\U0001F5FF"  # Simbol & Piktogram
        u"\U0001F680-\U0001F6FF"  # Transport & Simbol
        u"\U0001F1E0-\U0001F1FF"  # Bendera (sub-regional)
        u"\U00002702-\U000027B0"  # Simbol lain-lain
        u"\U000024C2-\U0001F251" 
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Menerapkan fungsi remove_emoji pada kolom 'text'
data_baru['wiI7pd'] = data_baru['wiI7pd'].apply(remove_emoji)
data_baru.reset_index(drop=True, inplace=True)
data_baru
```

```python
#hapus kolom yang memiliki missing value

data_baru = data_baru.dropna(axis=1)
data_baru
```

```python
folder_name = 'data_prepo'
file_path = os.path.join(folder_name, file_name)

# Membuat folder jika belum ada
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Menyimpan DataFrame ke format CSV di folder 'data_prepo'
data_baru.to_csv(file_path, index=False)

print(f"DataFrame telah disimpan di {file_path}")
```
